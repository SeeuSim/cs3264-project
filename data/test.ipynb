```python
filtered = df[df['station'] == 'CC1']

raw_columns = [
    'station',
    'start_time',
    'end_time',
    'crowd_level'
]
filtered = filtered[raw_columns]

filtered['crowd_level'] = filtered['crowd_level'].replace('na', 'l')

filtered['crowd_level'].unique()
```

```python
import re

def get_sorted_s_code(s: str):
    return re.sub(r'\d+$', lambda m: m.group().zfill(2), s)

sorted(list(df['station'].unique()), key=get_sorted_s_code)
```
